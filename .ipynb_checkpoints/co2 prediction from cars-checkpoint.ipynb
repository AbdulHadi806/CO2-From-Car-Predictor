```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error,mean_absolute_error 
from sklearn.model_selection import train_test_split


df = pd.read_csv('FuelConsumptionCo2.csv')
X = df[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_HWY']]
y = df['CO2EMISSIONS']
df
```

```python
X['ENGINESIZE']
# ploting data here
plt.figure(figsize=(13,5))
sns.lineplot(x=X['ENGINESIZE'], y=y)
plt.xlabel('ENGINESIZE')
plt.ylabel('CO2EMISSIONS')
plt.show()
```

```python
# spliting data in train and test with train_test_split 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train = X_train
y_train = y_train
X_test = X_test
y_test = y_test
plt.scatter(X_train['ENGINESIZE'], y_train, color='blue')
plt.xlabel('engine')
plt.ylabel('co2 emission')
plt.show()
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
predictCO2 = model.predict(X_test)
score = model.score(X_test, y_test)
print(score)
print(f'Sum of squared error (SSE): {np.sum((predictCO2 - y_test)**2)}') 
print(f'Mean squared error (MSE): {mean_squared_error(y_test, predictCO2)}') 
print(f'Sqrt of mean squared error (RMSE): {np.sqrt(mean_squared_error(y_test, predictCO2))}')

print(r2_score(predictCO2, y_test)) 
```

```python
plt.scatter(X_test['ENGINESIZE'], y_test, color='green')
plt.plot(X_test['ENGINESIZE'], model.coef_[0]*X_test['ENGINESIZE'] + model.intercept_, '-r')
plt.ylabel('CO2 emissions')
plt.xlabel('Engine')
plt.show()
```
